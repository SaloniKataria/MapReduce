```python
# Changing the directory of files
import os 
import re
from collections import OrderedDict
os.getcwd()
os.chdir('/Users/Isha/Downloads')

# Opening the text file as file 
file = open("mapreduce.txt", 'r')
data = file.read()
data
```

```python
#Dividing the data 
def div(i):
  lines = data.split('\n')
  return lines[:500],lines[500:]  

data1, data2 = div(data)
```

```python
# Create two different strings with 50:50 data recording the year, month and temp 
string1 = ""
string2 = ""
for x in data1:
  string1 = string1 + "," + x

for x in data2:
  string2 = string2 + "," + x

print(string1)
print(string2)
```

```python
# MAPPER REDUCER 
```

```python
#generating a list from the strings
import re
import collections

reducer1 = re.findall(r"\d+, \d+", string1)
reducer2 = re.findall(r"\d+, \d+", string2)

# Extract the year and temperature
def mapper(x):
  x = x[0:4: ] + x[6: :]
  return x

# Map the reducer values to the mapper function
map1 = map(mapper,reducer1)
list_1 = list(map(lambda x : (x.split(",")[0],x.split(",")[1]), list(map1)))
print(list_1)

map2 = map(mapper,reducer2)
list_2 = list(map(lambda x : (x.split(",")[0],x.split(",")[1]), list(map2)))
print(list_2)
```

```python
# Converting the list into dictionaries with years as key and temperature as values

def conv_dict(list):
    dict= {}
    for x,y in list:
        if x in dict.keys():
            dict[x].append(int(y))
        else:
            dict[x] = [int(y)]
    return dict

dict1 = conv_dict(list_1)
dict2 = conv_dict(list_2)
```

```python
#Combining the dictonaries created and sorting them by key for the whole dataset

union = dict1.keys() | dict2.keys()
combine = {key: dict1.get(key, []) + dict2.get(key, []) for key in union}
combine = OrderedDict(sorted(combine.items()))
combine.keys()
```

```python
#Partition Function
reducer1 = {}
reducer2 = {}

#All the months in year 2010 to 2015 are sent to Reducer1 and others to Reducer2
reducer1year= ['2010','2011','2012','2013','2014','2015']
reducer2year = ['2016', '2017', '2018', '2019', '2020']

for key in combine:
    if key[:4] in reducer1year:
        reducer1[key] = combine[key]
    else:
        reducer2[key] = combine[key]
```

```python
# REDUCER FUNCTION FOR MAX TEMP FOR EACH YEAR
```

```python
#Find the max temperature value in each key

reducer1_result = {}
for key in reducer1:
    max_value = max(reducer1[key])
    reducer1_result[key] = max_value

reducer2_result = {}
for key in reducer2:
    max_value = max(reducer2[key])
    reducer2_result[key] = max_value

print(reducer1_result)
print(reducer2_result)
```

```python
# IMPORTING THE RESULT TO OUTPUT.CSV
```

```python
results = {**reducer1_result, **reducer2_result}

import csv
dict = results
with open('output.csv', 'w') as f:
    for key in dict.keys():
        f.write("%s,%s\n"%(key,dict[key]))
```
